```python
import pandas as pd
import numpy as np
```

```python
#Retrieve envents from csv
def retrieve_meetup_events(csv):
    df = pd.read_csv(csv)
    clean_df = df.dropna(subset=["Event"], axis=0)
    events  = clean_df["Event"]
    return events
events = retrieve_meetup_events("datasets/MeetUp/events_Barcelona.csv")
```

```python
#Create users
RATES = 'Rates'
MOVIES = 'Movies'

class User():
    def __init__(self, name):
        self.name = name
        self.ratings = pd.DataFrame({MOVIES : [], RATES : []})
        
    def rate(self, movie, rating):
        try:
            if movie in self.ratings[MOVIES].tolist():
                self.ratings.loc[self.ratings[MOVIES] == movie, RATES] = rating
            else:
                self.ratings.append(pd.Series({MOVIES : movie, RATES : rating}))
        except TypeError:
            self.ratings = self.ratings.append(pd.Series({MOVIES : movie, RATES : rating}), ignore_index=True)
        
    
    def retrieve_rate(self, movie):
        return self.ratings.loc[self.ratings[MOVIES] == movie][RATES]
    
    def retrieve_movies_between(self, minrate, maxrate):
        return self.ratings[ (self.ratings[RATES]< maxrate) & (self.ratings[RATES] > minrate)]
    
    def retrieve_df(self):
        return self.ratings

        
```

```python
#Create Random User Database
np.random.seed(500)
users = []
user_names = []
for user_num in np.arange(10):
    user_name = "user_{}".format(user_num)
    user = User(user_name)
    for event in events:
        user.rate(event, np.random.uniform(0, 5))
    users.append(user)
    user_names.append(user_name)
```

```python
#Create Dataframe and write to csv
df = users[0].retrieve_df()
for user in users[1:]:
    df = pd.concat([df, user.retrieve_df()[RATES]], axis=1)
df.columns = ["Movies"] + user_names
df.to_csv("random_users.csv")
```
